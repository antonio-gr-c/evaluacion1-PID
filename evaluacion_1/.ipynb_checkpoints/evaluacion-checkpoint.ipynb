### COMANDO cut -d' ' -f 1-5

```python
!cut -d' ' -f 1-5 apache.log
```

!head ./Apache.log

El comando !head ./Apache.log te muestra las primeras 10 lneas del archivo Apache.log. En tu caso, las lneas muestran mensajes de un servidor Apache, incluyendo notificaciones y errores relacionados con LDAP, SSL, y problemas al crear ciertos componentes (channel.jni:jni y vm). En resumen:

LDAP est construido con el SDK de OpenLDAP, pero no tiene soporte SSL.
El mecanismo suEXEC est habilitado.
Hay problemas al intentar crear ciertos canales y mquinas virtuales (channel.jni y vm), lo que resulta en errores.
El comando !cut -d' ' -f 1-5 apache.log se utiliza para extraer las primeras 5 columnas de cada lnea en el archivo apache.log, usando el espacio (' ') como delimitador. As, te mostrara las partes que corresponden a la fecha, hora y nivel de mensaje (como notice o error). Esto es til para centrarse en las marcas de tiempo y los tipos de eventos sin ver toda la lnea del log.

cut: Sirve para cortar partes especficas de las lneas de un archivo.
-d' ': Usa el espacio como delimitador.
-f 1-5: Extrae los primeros 5 campos (columnas) de cada lnea.

```python
!wc -l Apache.log
```

```python
!grep -oP '\w{3} \d{4}' Apache.log | sort | uniq -c
```

#### NMERO DE EVENTOS POR CATEGORA (NOTICE, ERROR, WARN)

```python
!cat ./Apache.log | grep '\] \[' | cut -d ' ' -f 6 | sed 's/^\[//;s/\]$//' | sort | uniq -c
```
